```python
def convert_salary_to_usd(salary):

    conversion_rate = {'CA' : 0.75, '€' : 1.16, '£' : 1.27}
    
    if type(salary) == int:
        return salary

    elif 'CA' in salary:
        salary = int(re.sub('[^0-9]','', salary))
        salary = salary * conversion_rate['CA']
        salary = round(salary)
        return salary
    
    elif '€' in salary:
        salary = int(re.sub('[^0-9]','', salary))
        salary = salary * conversion_rate['€']
        salary = round(salary)
        return salary
    
    elif '£' in salary:
        salary = int(re.sub('[^0-9]','', salary))
        salary = salary * conversion_rate['£']
        salary = round(salary)
        return salary

    else:
        salary = int(re.sub('[^0-9]','', salary))
        return salary
    
    
```

```python
jobs['salary'] = jobs['salary'].apply(convert_salary_to_usd)
```

```python
common_locations = ['CA', 'NY', 'VA', 'IL', 'WA', 'OR' 'England', 'ON', 'GA', 'MA', 'TX', 'NC', 'MD', 'D.C.', 'NJ', 'MI', 'AZ', 'CO', 'Scotland', 'BC', 'QC', 'Ireland']

def process_location(location):
    location = location.split(",")
    try:
        state_province_country = location[1].strip()
        if state_province_country in common_locations:
            return state_province_country
        else:
            return "Other"
    except:
        return "Other"
    

      
```

```python
jobs['location_bin'] = jobs['location'].apply(process_location)
```
