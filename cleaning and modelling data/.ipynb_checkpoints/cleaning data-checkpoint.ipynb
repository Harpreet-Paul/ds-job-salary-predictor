```python
## function that converts all salary estimates from their native currency to USD

def convert_salary_to_usd(salary):
    
    ## mapping currencies to their USD conversion rate 

    conversion_rate = {'CA' : 0.75, '€' : 1.16, '£' : 1.27}
    
    ## some salaries are already type int and in USD so these are kept as is 
    
    if type(salary) == int:
        return salary
    
    ## all non-numeric characters in the salary are stripped, the remaining string is converted to type int and USD and then rounded to the nearest dollar. 

    elif 'CA' in salary:
        salary = int(re.sub('[^0-9]','', salary))
        salary = salary * conversion_rate['CA']
        salary = round(salary)
        return salary
    
    elif '€' in salary:
        salary = int(re.sub('[^0-9]','', salary))
        salary = salary * conversion_rate['€']
        salary = round(salary)
        return salary
    
    elif '£' in salary:
        salary = int(re.sub('[^0-9]','', salary))
        salary = salary * conversion_rate['£']
        salary = round(salary)
        return salary

    else:
        salary = int(re.sub('[^0-9]','', salary))
        return salary
    
    
```

```python
## applying the salary conversion function to the salary column 

jobs['salary'] = jobs['salary'].apply(convert_salary_to_usd)
```

### Cleaning Location

Most of the job locations were in a “city, state” format, e.g. “Austin, TX”. If the locations were kept at city level, the location feature would have been high cardinality. This would have posed two problems: (a) many of the locations at prediction time (when users use the app) would probably not be present in the training set and (b) the salary predictions would likely overfit to the job location. Thus, locations for job postings from the US or Canada were grouped by state or province and locations for job postings from England, Scotland or Ireland were grouped by country (since so few postings were from Europe). For example, “Austin, TX” was converted to simply “TX” and “London, England, United Kingdom” was converted to simply “England”. To account for unseen locations at the time of prediction, an “Other” category was created and locations that were not from one of 16 major US states, 3 major Canadian provinces or England/Scotland/Ireland were dumped into this category. 

```python
## Function that processes the location column. The full location is split at it's commas into a list of it's constituents.
## The second element of the list is the state/province/country. If the state/province/country is in the dictionary of major locations, then it is returned as the processed location.
## All other locations are processed as "Other"

common_locations = ['CA', 'NY', 'VA', 'IL', 'WA', 'OR' 'England', 'ON', 'GA', 'MA', 'TX', 'NC', 'MD', 'D.C.', 'NJ', 'MI', 'AZ', 'CO', 'Scotland', 'BC', 'QC', 'Ireland']

def process_location(location):
    location = location.split(",")
    try:
        state_province_country = location[1].strip()
        if state_province_country in common_locations:
            return state_province_country
        else:
            return "Other"
    except:
        return "Other"
    

      
```

```python
## Applying the location processing function to the location column. 

jobs['location_bin'] = jobs['location'].apply(process_location)
```

```python
## converting company type and industry columns to str because some of the entries are an integer.
## this step is needed so that the function in the next cell can be applied to these columns.

jobs['company_type'] = jobs['company_type'].astype(str)
jobs['industry'] = jobs['industry'].astype(str)
```

```python
## noticed that for some rows, the company type was scraped into the industry column.
## following function processes company type column by categorizing company type based on the presence of one of the dictionary keys. 
## also checks if company type was scraped into the industry column and converts the company type to value in the industry column if so. 

def fix_company_types(kind_of_company, industry):
    company_kinds = {"Public" : "Public", "Private" : "Private", "College / University" : "College / University", "Government" : "Government", "Non-profit Organisation" : "Non-profit Organisation", "Hospital" : "Hospital", "Subsidiary or Business Segment" : "Subsidiary or Business Segment", "Contract" : "Other"}
    for x in company_kinds:
        if (x in kind_of_company) or (x in industry):
            return company_kinds[x]
    return kind_of_company
```

```python
## applying the fix company types function to the company type and industry columns

jobs['company_type'] = jobs.apply(lambda x: fix_company_types(x['company_type'], x['industry']), axis=1)
```

```python
## Since some of the company rating values are in a weird string format that includes non-numeric characters, these need to be stripped away so that only numeric characters remain. 
## Function that will process the company ratings by stripping away non-numeric characters. 


def fix_company_ratings(company_rating):
    if type(company_rating) != str:
        return company_rating
    else:
        company_rating = re.sub('[^0-9.]','', company_rating)
        return company_rating
```

```python
## Applying the fix company ratings function to the company ratings column. 

jobs['company_rating'] = jobs['company_rating'].apply(fix_company_ratings)
```
